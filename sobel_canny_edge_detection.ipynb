```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded properly
if image is None:
    print("Error: Could not load image.")
    exit()
```

```python
# Sobel edge detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)  # Sobel in X direction
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)  # Sobel in Y direction
sobel_combined = np.sqrt(sobel_x**2 + sobel_y**2)  # Combine the two gradients
sobel_combined = np.uint8(sobel_combined)  # Convert back to uint8
```

```python
# Canny edge detection
canny_edges = cv2.Canny(image, 100, 200)  # Canny edge detection
```

```python
# Display the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Sobel Edge Detection')
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Canny Edge Detection')
plt.imshow(canny_edges, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
